```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 

%matplotlib inline
```

```python
df = pd.read_csv("Train Set - Train Set.csv")
df.set_index('Region_Name', inplace = True)
```

```python
df.head()
```

```python
df.tail()
```

```python
from statsmodels.tsa.api import SimpleExpSmoothing
import warnings
from sklearn.metrics import mean_absolute_percentage_error
warnings.filterwarnings("ignore")

total_regions = df.shape[0]
for i in range(total_regions):
    train = df.iloc[i][:-20]
    test = df.iloc[i][-20:]
    model=SimpleExpSmoothing(train)
    model_fit = model.fit()
    
    start=len(train)
    end=len(train)+len(test)-1
    pred = model_fit.predict(start = start, end = end)    
    
    print(mean_absolute_percentage_error(test, pred))
    
    
```

```python
from statsmodels.tsa.api import SimpleExpSmoothing
from sklearn.metrics import mean_absolute_percentage_error
import re

import warnings

warnings.filterwarnings("ignore")

total_regions = df.shape[0]

months = ["Month 73",
"Month 74",
"Month 75",
"Month 76",
"Month 77",
"Month 78",
"Month 79",
"Month 80",
"Month 81",
"Month 82",
"Month 83",
"Month 84",
"Month 85",
"Month 86",
"Month 87"]

final_preds = pd.DataFrame()

for i in range(total_regions):
    train = df.iloc[i]
    model=SimpleExpSmoothing(train)
    model_fit = model.fit()
    
    start = len(train) + 1
    end = len(train) + 15
    
    pred = model_fit.predict(start = start, end = end)
    final_preds[df.iloc[i].name] = pred
    
final_preds["Month"] = months
final_preds.set_index("Month", inplace=True)
final_preds
```

```python
final_preds.to_csv("test_data.csv")
```
